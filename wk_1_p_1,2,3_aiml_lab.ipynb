```
import pandas as pd
data = {


'people': ['sathvik', 'vineel', 'naviya', 'sneha'],
'age': [21, 24, 19, 433],
'height': [180, 130, 183, 159],
'weight': [55, 93, 59, 60]
}
df = pd.DataFrame(data)
print(df)
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['height'], s=100, color='green', label='Height')
plt.scatter(df['age'], df['weight'], s=100, color='yellow', label='Weight')
plt.title('Relationship between Age and Health Metrics')
plt.xlabel('Age')
plt.ylabel('Height / Weight')
plt.legend()
plt.grid(True)
plt.show()
```

```
plt.figure(figsize=(10, 8))
plt.bar(df['people'], df['height'], color='blue', label='Height')
plt.bar(df['people'], df['weight'], color='red', label='Weight')
plt.title('Comparison of Height and Weight among Individuals')
plt.xlabel('People')
plt.ylabel('Height / Weight')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```
plt.figure(figsize=(12, 6))
plt.hist(df['height'], bins=5, color='pink', alpha=0.7, label='Height')
plt.hist(df['weight'], bins=5, color='black', alpha=0.7, label='Weight')
plt.title('Distribution of Height and Weight')
plt.xlabel('Measurement')
plt.ylabel('Frequency')
```

```
import numpy as np
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
df=pd.read_csv('/content/drive/MyDrive/NFLX.csv')
df.head()
```

```
df['y']=df['Open']
df['ds']=df['Date']
df
```

```
train_df=df[(df['ds']>'2023-01-01') & (df['ds']<'2024-01-01')]
test_df=df[df['ds']>='2024-01-01']
```

```

print('Shape of train data',train_df.shape)
print('Shape of test data',test_df.shape)
```

```
model= Prophet()
model.fit(train_df)
```

```
future = model.make_future_dataframe(periods=13)
future.head()
```

```
future.shape
```

```
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```

forecast_plot = model.plot(forecast)
axes = forecast_plot.gca()
last_training_date = forecast['ds'].iloc[-12]
axes.axvline(x=last_training_date, color='red', linestyle='--', label='Training End')
test_df['Month'] = pd.to_datetime(test_df['ds'])
plt.plot(test_df['Month'], test_df['Open'],'ro', markersize=3, label='True Test Data')
plt.legend()
```
